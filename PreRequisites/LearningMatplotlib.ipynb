```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../input/NSEI-2.csv", usecols=['Open','High','Low','Close'],nrows=10)
#df['Date'] = pd.to_datetime(df['Date'])
df.astype(float)
plt.figure()
df.plot()
plt.show()

x = np.linspace(0,10,100)
#y = np.sin(x)
y = x ** 2

#Functional method of creating the plot
plt.plot(x,y,'g-')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My First Plot")
```

```python
# Subplots - plt.subplot(number of rows,number of columns, plot number)
plt.subplot(1,2,1)
plt.plot(x,y,'g-')
plt.subplot(1,2,2)
plt.plot(y,x,'r-')
```

```python
# Object Oriented Matplotlib
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("OOP X Label")
axes.set_ylabel("OOP Y Label")
axes.set_title("OOP Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'r')
axes1.set_title("Larger Plot")
axes2.plot(y,x,'b')
axes2.set_title("Smaller Plot")
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'r')
axes[1].plot(y,x,'g')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x ** 0.2, label='X Square')
ax.plot(x, x ** 0.3, label='X Cube')
ax.set_xlim([0,6])
ax.legend(loc=(0.1,0.1))
```

```python
df2 = pd.read_csv("../input/NSEI-2.csv", usecols=['Date','Open','High','Low','Close'],
                  nrows=10, index_col=None)
df2['Date'] =  pd.to_datetime(df2['Date'], format='%m/%d/%Y')
df2['Open'] = df2['Open'].astype(float)
df2['Close'] = df2['Close'].astype(float)
df2['High'] = df2['High'].astype(float)
df2['Low'] = df2['Low'].astype(float)

fig, ax = plt.subplots(nrows=1,ncols=2)
ax[0].plot(df2.Date, df2.Open,linewidth=2.5, linestyle="-", color="green",marker='o', markersize=10)
ax[1].plot(df2.Date, df2.Close,linewidth=2.5, linestyle="-", color="purple",marker='o', markersize=10)
#df2.plot()
plt.tight_layout()
fig.autofmt_xdate()
plt.show()
```
